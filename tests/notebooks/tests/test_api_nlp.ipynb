```python
import requests
import json

file = '../../docs/text4'
with open(file, 'r') as f:
    text = f.read()
    data = {'text': text, 'model': 'en'}
    headers = {'Content-type': 'application/json', 'Accept': 'text/plain'}
    url = 'http://localhost:5000/nlp'
    res = requests.post(url, data=json.dumps(data), headers=headers)
    print(res.text)
```
