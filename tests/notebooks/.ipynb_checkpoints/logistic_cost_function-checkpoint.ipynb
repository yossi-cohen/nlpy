### Sigmoid $\frac {1} {1+e^{-x}}$

```python
# plot sigmoid(x)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


# plot -log( h(x) )
x = np.linspace(-5, 5, 100)
plt.plot(x, sigmoid(x))
plt.title('Sigmoid')
plt.xlabel('$x$')
plt.xlim(-5, 5)
# plt.plot(x, sigmoid(x), label='$\frac {1} {1 + e^x}$')
# plt.xlabel('$x$')
# plt.legend()
plt.show()
```

## Logistic cost function

We cannot use the same cost function that we use for linear regression because the Logistic Function will cause the output to be wavy, causing many local optima. In other words, it will not be a convex function.

Instead, our cost function for logistic regression looks like:

$$J(\theta) = \frac {1}{m} \sum_{i=1}^m Cost\left(h_\theta\left(x^{(i)}\right), y^{(i)}\right)$$
$${
Cost\left(h_\theta\left(x\right), y\right) = -log\left(h_\theta\left(x\right)\right)$ \space 
    if \space y = 0 \\
Cost\left(h_\theta\left(x\right), y\right) = -log\left(1 - h_\theta\left(x\right)\right)$ \space 
    if \space  y = 1
}$$

<p>So we get:
$${
    {Cost}(h_\theta(x),y) = 0 \text{ if } h_\theta(x) = y \\
    {Cost}(h_\theta(x),y) \rightarrow \infty \text{ if } y = 0 \space and \space h_\theta(x) \rightarrow 1 \\
    {Cost}(h_\theta(x),y) \rightarrow \infty \text{ if } y = 1 \space and \space h_\theta(x) \rightarrow 0
}$$

<p>Note that $y = 0$ or $y = 1$ always, 
    <br>so we get a simplified (compressed) cost function:
<p>$$Cost\left(h_\theta\left(x\right), y\right) = -y \cdot log\left(h_\theta\left(x\right)\right)
    - (1-y) \cdot log\left(1 - h_\theta\left(x\right)\right)$$


```python
# plot log(z)
z = np.linspace(0.0001, 5.0)
plt.plot(z, np.log(z))
plt.title('log(z)')
plt.xlabel('$z$')
plt.show()
```

```python
# plot -log( h(x) )
x = np.linspace(-10.0, 10.0)
h = sigmoid(x)
plt.plot(h, -np.log(h))
plt.title('$-log(h(x))$ if y = 1')
plt.xlabel('$h(x)$')
plt.show()
```

```python
# plot -log( 1 - h(x) )
x = np.linspace(-10.0, 10.0)
h = sigmoid(x)
plt.plot(h, -np.log(1 - h))
plt.title('$-log(1-h(x))$ if y = 0')
plt.xlabel('$h(x)$')
plt.show()
```
